```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi

from skimage import io
from skimage.segmentation import watershed
from skimage.feature import peak_local_max
from skimage.filters import threshold_otsu

%matplotlib inline
```

### Preprocessing

Perform some image smoothing to reduce image noise

```python
image = io.imread("../images/nuclei-dapi.tif")
```

### Generate candidates for markers

```python
threshold = threshold_otsu(image)
print(threshold) 
binary = ~(image <= threshold)
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(binary, cmap='gray')
plt.show()
```

```python
distance = ndi.distance_transform_edt(binary)
```

#### Watershed Parameters: 

- threshold_m: We set this threshold to place the markers near the peaks of the distance transform
- footprintsize: Size of the local neibourhood when finding local maxima 

```python
threshold_m = threshold_otsu(distance)
footprintsize = 21
```

```python
coords = peak_local_max(distance, threshold_abs=threshold_m, 
                        footprint=np.ones((footprintsize, footprintsize)), labels=image)
mask = np.zeros(distance.shape, dtype=bool)
print(coords.shape)
mask[tuple(coords.T)] = True
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
plt.subplot(1,2,1)
plt.imshow(distance, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(mask, cmap='gray')
plt.show()
```

### Use the watershed algorithm to generate a label image

```python
markers, _ = ndi.label(mask)
labels1 = watershed(distance, markers, mask=image, compactness=0)
labels2 = watershed(distance, markers, mask=image, compactness=.01)
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
plt.subplot(1,2,1)
plt.imshow(labels1)
plt.subplot(1,2,2)
plt.imshow(labels2)
plt.show()
```
